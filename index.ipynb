## Dati geografici 



Esistono differenti e numerose codifiche per esprimere i dati spaziali. 



Il Sistema di Riferimento Globale (ETRS89)  viene indicato, a volte, con l’acronimo “WGS84”. 

Per approfondire si consiglia di consultare [questo link](https://en.wikipedia.org/wiki/Spatial_reference_system)

Le sigle SRS (spatial reference system) e CRS(coordinate reference system) sono sinonimi. 

I dati geografici per poter essere visualizzati dai tool che la piattaforma mette a disposizione devono rispettare la codifica WGS84.

Nel dataset delle strutture ricettive **non** è indicato il tipo di formato usato.

In questo caso è necessario capire in quale formato sono espressi e convertire in WSG84

Per effettuare "operazioni geografiche" sui dataset abbiamo a disposizione la libreria [**geopandas**](http://geopandas.org/gallery/create_geopandas_from_pandas.html#sphx-glr-gallery-create-geopandas-from-pandas-py) 


[Link utile per convertire coordinate](http://www.geoin.it/coordinate_converter/)



http://spatialreference.org/


EPSG SRID

4326 WGS 84 Geografico 

Note: mondiale; usato dagli strumenti GPS



3003 Monte Mario / Italia zona 1 

3004 Monte Mario / Italia zona 2 

Note: obsoleto (1940), ma ancora comunemente usato


23032 ED50 / UTM zone 32N 

23033 ED50 / UTM zone 33N

Note: superati e raramente utilizzato: European Datum 1950


32632 WGS 84 / UTM zone 32N

32633 WGS 84 / UTM zone 33N

Note: WGS84, adotta la proiezione piana UTM

### 1 - Creare un GEOdataframe 
partendo dal dataframe Strutture

```python
import geopandas as gpd
```

```python
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

Un GeoDataFrame ha bisogno di contenere un oggetto "shapely",  

dobbiamo quindi creare una nuova colonna con i dati geografici che abbiamo nel csv, una tupla di Longitude e Latitude

```python
strutture['geometry'] = list(zip(strutture.COORDX, strutture.COORDY))
```

```python
strutture['geometry'].head(5)
```

### Trasformiamo le tuple in oggetti "Point"

```python
strutture['geometry'] = strutture['geometry'].apply(Point)
```

```python
strutture['geometry'].head(5)
```

### Adesso possiamo creare il geodataframe
Poichè i dati provengono da un CSV è necessario sapere quale  datum/sphereoid/proiezione è stata usata.

Questa è un'informaziona da indicare nei metadati ma che qui è **assente (!)**

Possiamo  andare per tentativi facendoci aiutare dal nostro intuito (o esperienza...)

```python
strutture_geo = gpd.GeoDataFrame(strutture, geometry=strutture.geometry)
```

```python
strutture_geo.head(4)
```

### Mancano i metadati (!) - come ricoscere il sistema di riferimento geografico usato?

#### Proviamo ad individuare il sistema di coordinate individuato

http://epsg.io

sembrerebbe essere epsg:3004 Monte Mario / Italia zona 2


```python
strutture_geo.crs = {'init': 'epsg:3004'}
```

```python
strutture_geo.plot()
```

### Abbiamo visto che molti record non possiedono i campi COORDX e COORDY valorizzati.
Per evitare rappresentazioni che includono questi valori insignificanti ed errati, 

attribuiamo un valore scelto a caso di localizzato appena fuori Italia (1956886, 5128102)

```python
strutture_geo.COORDX.isnull().sum()
```

```python
strutture_geo.COORDY.isnull().sum()
```

```python
#valorizzi i campi con i valori di default
strutture_geo.COORDX.fillna(1956886, inplace=True)
strutture_geo.COORDY.fillna(5128102, inplace=True)
#ricreo la colonna geometry
strutture_geo['geometry'] = list(zip(strutture_geo.COORDX, strutture_geo.COORDY))
strutture_geo['geometry'] = strutture_geo['geometry'].apply(Point)
```

### Disegniamo la mappa delle regioni d'Italia 

Puoi scaricare i file geografici da ISTAT a questo link

[BASI TERRITORIALI E VARIABILI CENSUARIE](https://www.istat.it/it/archivio/104317)


Per comodità trovi il file "ISTAT_regioni.geojson" con le regioni italiane. nella cartella ./data 

```python
italia = gpd.read_file("./data/ISTAT_regioni.geojson")
italia.plot()
```

```python
print (italia.crs)
```

### Convertiamo la rappresentazione dell'Italia nel formato Monte Mario/Italy zone 2 

```python
italia.crs = {'init': 'epsg:3004'}
italia.plot()
```

```python
#primo layer
ax = italia.plot(color='red',  edgecolor='white', figsize=(15,5))
#secondo layer
strutture_geo.plot(ax=ax)
ax.axis('off')
```

### Anche se ci sono molti valori che sono a null, sembrerebbe che abbiamo indovinato la giusta codifica.

Adesso possiamo convertire così in quella richiesta ovvero WGS84 

```python
strutture_geo = strutture_geo.to_crs({'init': 'epsg:4326'})
```

### Creiamo le due nuove colonne Latitudine e Longitudine

```python
strutture_geo["Longitudine"]= strutture_geo.geometry.x

strutture_geo["Latitudine"]= strutture_geo.geometry.y
```

```python
strutture_geo
```

### Uniamo le info deldataset *Strutture* e del dataset *Codifiche*

Attraverso il comando "merge" offerto dalla libreria Pandas, in modo da avere un unico dataframe contenente anche le decodifiche/informazioni racchiuse nel dataset delle codifiche

[Consulta la documentazione](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html)

```python
strutture_codifiche = strutture.merge(codifiche, how='inner',left_on='APT', right_on='CODICE')
```

```python
strutture_codifiche.head(3).T
```

### Check su dataframe finale *strutture_codifiche*
Il numero di righe del dataframe risultante deve avere lo stesso numero di righe di quello iniziale a patto che ci sia integrità tra la tabella dei fatti (strutture) e quella della dimensione (codifiche)

```python
strutture_codifiche.shape
```

```python
strutture.shape
```

### Eliminiamo i campi non necessari 

C'è una duplicazione di informazione: i campi  "APT" e "CODICE"  possiedono la stessa info.

Possiamo quindi eliminare il campo "codice" e anche i campi "_id_y" e "_id_x" che sono identificativi di cui non conosciamo l'utilità. 

Inoltre possiamo eliminare i campi geografici originali e creare due nuove colonne "Latitudine" e "Longitudine" i cui valori li estraiamo dal campo "geometry"

```python
strutture_codifiche.drop(['_id_y','_id_x','CODICE','ID', 'COORDX','COORDY'], axis=1)
```

```python
strutture_codifiche.groupby(strutture_codifiche.TIPOLOGIA).size().plot.bar(title="strutture per tipologia",figsize=(10,5))
```

### Salviamo il dataset finale
Per caricare sulla piattaforma DAF ho bisogno di un sample del dataset e dell'inter dataset in formato csv.
La libreria Pandas ci permette di farlo agilmente in un unico comando:

In particolare utilizziamo il comando **nome_dataframe.to_csv()**

Si possono specificare molte opzioni ad esempio sul *separatore* da usare, il tipo di *encoding*...

Il parametro **index=False** fa sì che non venga esportata anche la prima colonna contenente l'indice del dataframe.

[Consulta la documentazione](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_csv.html)

### DAF invio dataset con SFTP

Dopo aver metadato il dataset, è necessario inviare nella cartella specificata nella scheda informativa il dataset completo.

Puoi utilizzare comodamente alcuni tool come [Filezilla](https://filezilla-project.org/) oppure farlo attraverso alcune righe di codice.

Dovrai inserire il tuo utente, la tua password e autenticarti via SFTP a daf.teamdigitale.it (porta 22)

Importiamo il modulo **pysftp** che ci permette di effettuare il trasferimento via SFTP

[Consulta la documentazione](http://pysftp.readthedocs.io/en/release_0.2.8/cookbook.html)

```python
import pysftp
```

```python
sftp_host = "daf.teamdigitale.it"
sftp_user = "user"
sftp_password = "secret"


local_path = "./data/strutture_umbria.csv"
sftp_path = "test/2"

cnopts = pysftp.CnOpts()
cnopts.hostkeys = None
cnopts.compression = True

with pysftp.Connection(host=sftp_host, 
                       username=sftp_user,
                       password=sftp_password, 
                       cnopts=cnopts) as sftp:
    sftp.chdir(sftp_path)
    sftp.put(local_path, confirm=True)
```
